```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
```

```python
# we create 40 seperate points
'''
n_samples: Number of points
centers: Either the number of centers, or a matrix of the chosen centers
random_state: 
'''
x, y = make_blobs(n_samples=40, centers=2, random_state=20)
```

```python
x[0:5], y[0:5]
```

```python
clf = svm.SVC(kernel='linear', C=1)
clf.fit(x, y)
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.show()
# blue is crocodile and red is aligator
```

```python
clf.predict([[ 1.71842122,  9.52190711]])
```

```python
# plotting using support vector machine
clf = svm.SVC(kernel='linear', C=1)
clf.fit(x, y)
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decesion boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(xlim[0], xlim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

#plot decesion boundry and margins
ax.contour(XX, YY, Z, colors='k', 
           levels=[-1, 0, 1], alpha=0.5, 
           linestyles=['--', '-', '--'])

# plot support vector
ax.scatter(clf.support_vectors_[:, 0],
           clf.support_vectors_[:, 1], s=100, 
           linewidth=1, facecolors='none')
plt.show()
```
