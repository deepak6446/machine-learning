```python
def get_unique(tr, te, col):
    unique = list(set(np.append(test[col].unique(),train[col].unique())))
    return unique
```

```python
train = train.applymap(lambda s:s.lower() if type(s) == str else s)
test = test.applymap(lambda s:s.lower() if type(s) == str else s)
```

```python
get_unique(train, test, 'Outlet_Size')
train['Outlet_Size'].unique()
```

```python
def clean_Outlet_Size(data):
    data['Outlet_Size'] = data['Outlet_Size'].fillna("Not defined")
    return data

train = clean_Outlet_Size(train)
test = clean_Outlet_Size(test)
train['Outlet_Size'].unique()
```

```python
def clean_Item_Weight(data):
    data['Item_Weight'].fillna(0, inplace=True)    
    return data
train = clean_Item_Weight(train)
test = clean_Item_Weight(test)
```

```python
train.isna().sum(axis=0)
```

```python
def clean_Item_Identifier(data):
    del data['Item_Identifier']
    return data
train = clean_Item_Identifier(train)
test = clean_Item_Identifier(test)
```

```python
get_unique(train, test, 'Item_Fat_Content')
```

```python
def clean_Item_Fat_Content(data):
    data['Item_Fat_Content'] = data['Item_Fat_Content'].str.replace('low fat', 'lf')
    data['Item_Fat_Content'] = data['Item_Fat_Content'].str.replace('regular', 'reg')
    return data
train = clean_Item_Fat_Content(train)
test = clean_Item_Fat_Content(test)
```

```python
get_unique(train, test, 'Item_Type')
```

```python
def clean_Item_Type(data):
    data['Item_Type'] = data['Item_Type'].str.replace('breads', 'starchy foods')
    data['Item_Type'] = data['Item_Type'].str.replace('frozen foods', 'canned')
    data['Item_Type'] = data['Item_Type'].str.replace('meat', 'nonveg')
    data['Item_Type'] = data['Item_Type'].str.replace('seafood', 'nonveg')
    return data
train = clean_Item_Type(train)
test = clean_Item_Type(test)
```

```python
get_unique(train, test, 'Outlet_Identifier')
```

```python
get_unique(train, test, 'Outlet_Establishment_Year')
```

```python
def clean_Outlet_Establishment_Year(data):
    data['Outlet_Establishment_Year'] = 2019 - data['Outlet_Establishment_Year'] 
    return data
```

```python
train = clean_Outlet_Establishment_Year(train)
test = clean_Outlet_Establishment_Year(test)
get_unique(train, test, 'Outlet_Establishment_Year')
```

```python
y = train['Item_Outlet_Sales']
del train['Item_Outlet_Sales']
```

```python
def getDummies(data, columns):
    data = pd.get_dummies(data, columns=columns)
    return data
```

```python
train = getDummies(train, ['Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', ])
train.head()
```

```python
test = getDummies(test, ['Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', ])
test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
columns = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year'] 
temp = pd.concat([train, test], axis=0)
scaler = MinMaxScaler()
scaler = scaler.fit(temp[columns])
def minMaxScaler(data, columns):
    data[columns] = scaler.transform(data[columns])
    return data
```

```python
train = minMaxScaler(train, columns)
train.head()
```

```python
test = minMaxScaler(test, columns)
```

```python
test.head()
```

```python
y_scaler = MinMaxScaler()
y = np.array(y)
y_w = y[:]
# test_y = y_w[0:1000]
y = y.reshape(-1, 1)
y_scaler = y_scaler.fit(y)
y = y_scaler.transform(y)
```

```python
x = np.array(train)
x_pred_test = np.array(test)
# test_x = x[0:1000]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.30, random_state=42)
```

```python
print('x train', x_train.shape)
print('y train', y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Dense
from keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(patience=5)

model = Sequential()
model.add(Dense(100, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='relu'))
model.compile(loss='mean_squared_error', optimizer='adam',  metrics=['mse'])

history = model.fit(x, y, validation_split=0.2, epochs=1000, callbacks=[early_stopping_monitor])
```

```python
# loss: 0.0067 - val_loss: 0.0073  model --> 100-->10-->1
index = 8
x_test_pre = np.array(x[index]).reshape(1, -1)
```

```python
y_test = model.predict(x_test_pre)
y_test = y_test.reshape(-1, 1)
print("predicted:", y_test)
print("actual: ", y_w[index])
print(y_scaler.inverse_transform(y_test)[0][0])
```
