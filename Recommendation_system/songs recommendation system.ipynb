```python
import pandas as pd
from sklearn.model_selection import train_test_split
import Recommenders as Recommenders
#import Evaluation as Evaluation
```

<h3>Loading song data</h3>
<li>1st data set contains user_id, sond_id and the number of times user listened to that song.
<li>2nd file contains song meta data

```python
song_df_1 = pd.read_table('/home/deepak/Desktop/deepWork/machineLearning/dataset/songData/10000.txt',header=None)
song_df_1.columns = ['user_id', 'song_id', 'listen_count']

song_df_2 =  pd.read_csv('/home/deepak/Desktop/deepWork/machineLearning/dataset/songData/song_data.csv')
```

```python
song_df_1.head()
```

```python
song_df_2.head()
```

```python
#Merge the two dataframes above to create input dataframe for recommender systems
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on="song_id", how="left")
```

```python
song_df.head()
```

```python
song_df.shape
```

```python
#Merge song title and artist_name columns to make a merged column
song_df['song'] = song_df['title'] + " - " + song_df['artist_name']
```

```python
song_df.head()
```

<h2> Getting the most popular songs in the dataset</h2>

```python
song_grouped = song_df.groupby(['song']).agg({'listen_count': 'count'}).reset_index()
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage']  = song_grouped['listen_count'].div(grouped_sum)*100
song_grouped.sort_values(['listen_count', 'song'], ascending = [0,1])
```

<h1> Getting number of unique users</h1>

```python
unique = song_df['user_id'].unique()
len(unique)
```

<h1>Getting number of unique songs in dataset</h1>

```python
songs = song_df['song'].unique()
len(songs)
```

<h1>Splitting data into train and test</h1>


```python
train_data, test_data = train_test_split(song_df, test_size = 0.20, random_state=0)
print(train_data.head(5))
```

<h1>create popularity based recommendation system</h1>

```python
popModel = Recommenders.popularity_recommender_py()
popModel.create(train_data, 'user_id', 'song')
```

```python
user_id  = unique[4]
popModel.recommend(user_id)
```

```python
colabModel = Recommenders.item_similarity_recommender_py()
colabModel.create(train_data, 'user_id', 'song')
```

Precision: What proportion of positive identifications was actually correct
eg.the percentage of emails flagged as spam that were correctly classified
recall: What proportion of actual positives was identified correctly?
the percentage of actual spam emails that were correctly classified ?
